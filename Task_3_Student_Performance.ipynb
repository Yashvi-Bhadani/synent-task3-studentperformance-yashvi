```python
import pandas as pd
# Create numeric target column
df['Performance_Category_Num'] = df['Performance Category'].map({
    'Low': 0,
    'Medium': 1,
    'High': 2
})

# Create copy
df_corr = df.copy()
df_corr['Performance_Category_Num'] = df_corr['Performance_Category_Num'].astype(int)

# Convert all categorical columns to numeric codes
for col in df_corr.select_dtypes(include='object').columns:
    df_corr[col] = df_corr[col].astype('category').cat.codes

# Correlation matrix
corr = df_corr.corr(numeric_only=True)

# Correlation with target
corr_target = corr['Performance_Category_Num'].sort_values(ascending=False)

print(corr_target)
# df_corr.dtypes
```

```python
import matplotlib.pyplot as plt

# Remove target itself
corr_target = corr_target.drop('Performance_Category_Num')

# Sort values
corr_target = corr_target.sort_values()

# Plot
plt.figure(figsize=(10, 8))
corr_target.plot(kind='barh')

plt.title('Correlation with Performance Category')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')

plt.tight_layout()
plt.show()
```
